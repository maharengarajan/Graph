```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
#generate dataset using randn fun
np.random.randn(1000)
```

```python
#convert array into dataset
df = pd.DataFrame(np.random.randn(1000), columns=['data'])
df
```

```python
#assigning date format in indexes
df = pd.DataFrame(np.random.randn(1000), columns=['data'],index=pd.date_range('2023/08/09',periods=1000))
df
```

```python
df.plot()
```

```python
#increasing size of the plot
df.plot(figsize=(20,10))
```
